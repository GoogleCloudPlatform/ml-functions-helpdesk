```python
import google.datalab.bigquery as bq
```

```python
def get_distinct_values(column_name):
    sql = """
SELECT
  {0},
  COUNT(1) AS num_tickets,
  AVG(resolutiontime) AS avg_rt
FROM
  `mam-cloud.helpdesk.tickets`
GROUP BY
  {0}
ORDER BY {0}
    """.format(column_name)
    return bq.Query(sql).execute().result().to_dataframe()
```

```python
df = get_distinct_values('impact')
df.plot(x='impact', y='num_tickets', logy=True, kind='bar')
df.plot(x='impact', y='avg_rt', kind='bar')
```
